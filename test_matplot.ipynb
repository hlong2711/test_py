### Simple plot


```python
# draw single line with 4 y-points
import matplotlib.pyplot as plt
import numpy as np

# x = [1, 2, 3, 4]
y = np.array([7, 4, 10, 8])  # default x is array start from 0
plt.plot(y, "o:r", ms=10)  # maker,line,color
plt.show()
```
